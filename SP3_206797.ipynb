```python
import ffmpy
import ffmpeg
import os
```

```python
## Exercici 1 
def convert(name, output, mode):
    if (mode == "VP8"):
        ff = ffmpy.FFmpeg(
            inputs={''+str(name)+'': None},
            outputs={''+str(output)+'': '-c:v libvpx -b:v 2M -c:a libvorbis'})
        ff.cmd
        ff.run()
    elif (mode == "VP9"):
        ff = ffmpy.FFmpeg(
            inputs={''+str(name)+'': None},
            outputs={''+str(output)+'': '-c:v libvpx-vp9 -b:v 2M '})
        ff.cmd
        ff.run()
    elif (mode == "h265"):
        ff = ffmpy.FFmpeg(
            inputs={''+str(name)+'': None},
            outputs={''+str(output)+'': '-c:v libx265 -crf 26 \
                                         -preset fast -c:a aac -b:a 128k'})
        ff.cmd
        ff.run()
    elif (mode == "AV1"):
        ff = ffmpy.FFmpeg(
            inputs={''+str(name)+'': None},
            outputs={''+str(output)+'': '-c:v libaom-av1 \
                                         -crf 30 -b:v 0'})
        ff.cmd
        ff.run()

## Exercici 2

def comparison(video1,video2,output): 
        ff = ffmpy.FFmpeg(
            inputs={''+str(video1)+'': None,
                    ''+str(video2)+'': None},
            outputs={''+str(output)+'': '-filter_complex hstack'})
        ff.cmd
        ff.run()
    
```
